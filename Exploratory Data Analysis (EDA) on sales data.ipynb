```python
# Import necessary libraries

from scipy import stats

# Calculate central tendency for Sales and Profit
for col in ['Sales', 'Profit']:
    mean_val = df[col].mean()
    median_val = df[col].median()
    mode_val = df[col].mode()[0]  # Take the first mode in case of multiple
    print(f"--- {col} ---")
    print(f"Mean: {mean_val:.2f}")
    print(f"Median: {median_val:.2f}")
    print(f"Mode: {mode_val:.2f}")
    print("-" * 30)
```

```python

plt.figure(figsize=(12, 5))

# Sales histogram
plt.subplot(1, 2, 1)
plt.hist(df['Sales'], bins=30, color='skyblue', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')

# Profit histogram
plt.subplot(1, 2, 2)
plt.hist(df['Profit'], bins=30, color='lightcoral', edgecolor='black')
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Calculate Measures of Spread for Sales and Profit

for col in ['Sales', 'Profit']:
    std_dev = df[col].std()
    variance = df[col].var()
    data_range = df[col].max() - df[col].min()
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    print(f"--- {col} ---")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Variance: {variance:.2f}")
    print(f"Range: {data_range:.2f}")
    print(f"Interquartile Range (IQR): {IQR:.2f}")
    print("-" * 35)
```

```python
# Function to calculate IQR-based outliers
def iqr_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    print(f"--- {column} ---")
    print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
    print(f"Number of Outliers: {outliers.shape[0]}")
    print("-" * 40)
    return outliers

# Detect outliers in Sales and Profit
outliers_sales = iqr_outliers('Sales')
outliers_profit = iqr_outliers('Profit')
```

```python
# Compute Z-scores for Sales and Profit
df['Z_Sales'] = stats.zscore(df['Sales'])
df['Z_Profit'] = stats.zscore(df['Profit'])

# Identify outliers (|Z| > 3)
outliers_sales = df[np.abs(df['Z_Sales']) > 3]
outliers_profit = df[np.abs(df['Z_Profit']) > 3]

# Display results
print(f"Outliers in Sales (|Z| > 3): {outliers_sales.shape[0]}")
print(f"Outliers in Profit (|Z| > 3): {outliers_profit.shape[0]}")
print("\nExample of Profit Outliers:")
print(outliers_profit[['Sales', 'Profit', 'Z_Profit']].head())
```

```python
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = df[(df['Profit'] >= lower) & (df['Profit'] <= upper)]
print(f"Original rows: {df.shape[0]}, After removal: {df_no_outliers.shape[0]}")
```

```python
from scipy.stats.mstats import winsorize

# Cap Sales values at 5th and 95th percentiles
df['Sales_Winsorized'] = winsorize(df['Sales'], limits=[0.05, 0.05])
```

```python
# Apply log transformation
df['Log_Sales'] = np.log1p(df['Sales'])  # log1p avoids log(0)
```

```python
# Function to remove outliers using the IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    filtered_df = df[(df[column] >= lower) & (df[column] <= upper)]
    return filtered_df

# Apply the function to Sales and Profit columns
df_no_outliers = remove_outliers_iqr(df, 'Sales')
df_no_outliers = remove_outliers_iqr(df_no_outliers, 'Profit')

print(f"Original dataset size: {df.shape[0]}")
print(f"After removing outliers: {df_no_outliers.shape[0]}")
```

```python
# compare boxplots before and after removing outliers:

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.boxplot(df['Profit'])
plt.title('Before Removing Outliers')

plt.subplot(1, 2, 2)
plt.boxplot(df_no_outliers['Profit'])
plt.title('After Removing Outliers')

plt.tight_layout()
plt.show()
```
